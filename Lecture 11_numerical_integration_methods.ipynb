## Trapezoidal Rule

### Algorithm:
1. Divide the interval [a, b] into n subintervals of equal width h = (b - a)/n.
2. Compute the sum: Integral ≈ (h/2) * [f(x₀) + 2f(x₁) + 2f(x₂) + ... + 2f(xₙ₋₁) + f(xₙ)].
3. Return the result.


```python
# Your code here
import matplotlib.pyplot as plt
import numpy as np

# Declare interval & subinterval
a = 0
b = np.pi
n = 100
h = (b-a)/n

# Declare x
x = np.linspace(a, b, n)

# Decalre function
def f(x):
    return np.sin(x)
y = f(x)

# Apply the formula
trap_integral = (h/2) * (y[0] + 2 * np.sum(y[1:-1]))

print(trap_integral)
```
